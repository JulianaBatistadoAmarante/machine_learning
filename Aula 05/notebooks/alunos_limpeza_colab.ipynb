### Pergunta para reflexão
Devemos remover alunos com nota nula ou preencher a nota? Justifique.


```python
media_nota = df['Nota'].mean()
df['Nota'] = df['Nota'].fillna(media_nota)
```

## 5. Registros duplicados


```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

## 6. Padronizando textos


```python
df['Nome'] = df['Nome'].astype(str).str.strip().str.title()
df['Curso'] = df['Curso'].astype(str).str.strip().str.title()
df['Curso'] = df['Curso'].replace({'Banco De Dados':'Banco de Dados'})
df['Curso'].unique()
```

## 7. Corrigindo valores inválidos


```python
df[df['Idade'] < 0]
```

```python
idade_mediana = int(df[df['Idade'] > 0]['Idade'].median())
df.loc[df['Idade'] < 0, 'Idade'] = idade_mediana
```

```python
df.loc[df['Nota'] > 10, 'Nota'] = 10
df.loc[df['Frequencia'] > 100, 'Frequencia'] = 100
```

## 8. Convertendo datas


```python
df['Data_Matricula'] = pd.to_datetime(df['Data_Matricula'], dayfirst=True, errors='coerce')
df['Data_Matricula'].head()
```

## 9. Preparando para Machine Learning


```python
X = df[['Idade', 'Nota', 'Frequencia', 'Atividades_Entregues']]
y = df['Aprovado']
X.head(), y.head()
```

## 10. Salvando o dataset tratado


```python
df.to_csv('alunos_tratado.csv', index=False)
```
